## Step 1 - Create and Display

```python
# import the necessary libraries 
import solution_registry as sr
# Establish the solution Name 
solution_name = 'solution_registry'
# Instantiate the lists 
solution_registry = sr.solution_registry_list() 
df = solution_registry.get_dataframe()
df.head(100)
```

## Step 2 - Persist for backup and editing purposes   

```python
def persist_list(): 
    import configparser, webbrowser 
    config = configparser.ConfigParser()
    cfg = config.read('config.ini')  

    # save and display formatted html  
    save_file_name = config.get('global_infrastructure', 'infrastructure_root_directory') + '/'  + \
    config.get('global_infrastructure', 'html_subdirectory') + '/'  +  solution_name +  \
    config.get('global_infrastructure', 'html_extension')
    print(save_file_name)
    solution_registry.write_html_file(save_file_name)
    # webbrowser.open_new_tab(save_file_name) 

    # save excel   
    save_file_name = config.get('global_infrastructure', 'infrastructure_root_directory') + '/'  + \
    config.get('global_infrastructure', 'excel_subdirectory') + '/'  +  solution_name +  \
    config.get('global_infrastructure', 'excel_extension')
    df.to_excel(save_file_name, index=False)
```

```python
persist_list()
```
